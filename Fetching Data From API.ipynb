```python
import requests
import csv
from datetime import datetime, timedelta


url = "https://api.weatherbit.io/v2.0/history/airquality"


api_key = "99657a02c78e479aa9b1bf411ebf02fb"


start_date = datetime.strptime("2023-01-01", "%Y-%m-%d")
end_date = datetime.strptime("2023-08-22", "%Y-%m-%d")
delta = timedelta(days=30)
csv_file_path = "air_quality_data.csv"

with open(csv_file_path, mode='w', newline='') as csv_file:
    fieldnames = []
    writer = None

    current_date = start_date
    while current_date <= end_date:
        next_date = current_date + delta if current_date + delta <= end_date else end_date
        parameters = {
            "city": "Hyderabad",
            "country": "India",
            "start_date": current_date.strftime("%Y-%m-%d"),
            "end_date": next_date.strftime("%Y-%m-%d"),
            "key": api_key
        }


        response = requests.get(url, params=parameters)


        if response.status_code == 200:
            data = response.json()
            air_quality_data = data.get('data')


            if writer is None:
                fieldnames = air_quality_data[0].keys()
                writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
                writer.writeheader()


            for row in air_quality_data:
                writer.writerow(row)

            print(f"Data for {current_date.strftime('%Y-%m-%d')} to {next_date.strftime('%Y-%m-%d')} written to CSV.")

        else:
            print("Error:", response.status_code)

        current_date = next_date + timedelta(days=1)
```
